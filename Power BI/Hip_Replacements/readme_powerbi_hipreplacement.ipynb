# Length of Stay and Cost of Discharge Analysis of Hip Replacements
Purpose:\
Analyze efficiency of inpatient stays for elective hip replacements in New York hospitals through length of stay (LOS) and Cost of Discharge. 
 
Dataset:\
Actual New York state-wide hospital discharges in 2016 for elective hip replacement surgery patients. All PHI and PII has been removed. 
 
Key Power BI Concepts:
* Data import, profiling, cleaning, transformations
* Data modeling
* Measure creation, DAX, Key Influencers analysis
 
Takeaways:
* New York City area LOS is 1.68 higher than all other service areas and contributes to 27% of the data
* New York City area cost of discharge is $15K higher than all other service areas and contributes to 27% of the data. 
* Hospitals with a large surgical program size (400-599) had an average cost of discharge of $5,400 less than other hospitals. 
 
Further Analysis:
